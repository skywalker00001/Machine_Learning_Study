```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f1(t):
    M = C * np.exp(-a1 * t)
    return M

C = 1
a1 = 0.1

t = np.linspace(0, 15, 300)
M = f1(t) 
```

```python
fig = plt.figure(facecolor="white")
plt.plot(t, M, 'r-', linewidth=2)
plt.title("Decay")
plt.show()
```
